# 3D Frequency-Domain Boundary Element Method (BEM)

This tutorial demonstrates the implementation of a 3D Boundary Element Method (BEM) to analyze the interaction between regular linear water waves and submerged porous structures.

The formulation is based on:
- Linear potential flow theory
- Time-harmonic (frequency-domain) analysis
- Boundary integral equation formulation

The objective is to compute:
- Wave transmission and reflection

```python
# Variables of interest

# Water depth
DEPTH         = 0.50;

RADIUS        = 0.30;
WIDTH_CHANNEL = 0.25*DEPTH;

LX = 3.00*DEPTH;

NUM_OF_REG = 1;

# Number of structures
N_STRUCT = 1;

gravity = 9.8056
rho     = 999.1033
```

```python
# GMSH mesh
SEMI_CIRCULAR(DEPTH,RADIUS,WIDTH_CHANNEL,LX)
```

```python
# NUMBER OF FREQUENCIES
ttmax = 1

# WAVE PERIODS
kh = np.linspace(3.00, 3.00, ttmax)
# WAVE HEIGHT
HL = 0.01
```

```python
# Process the outer region
MESH_0_POS, MESH_0_QUADS9 = PROCESS_MESH("SEMI_CIRCULAR.msh")
```

```python
# Number of nodes per quad element
NCONEC = 9;
```

```python
# PRE-ASSIGN
N  = [0] * NUM_OF_REG
NE = [0] * NUM_OF_REG
```

```python
N[0] ,  NE[0] ,  KCONEC0 = COORDINATES(MESH_0_POS , MESH_0_QUADS9)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(
    MESH_0_POS[:,0],
    MESH_0_POS[:,1],
    MESH_0_POS[:,2],
    c='k',
    s=1
)

#ax.grid(True)
ax.set_box_aspect([7.5,0.5,1])  # axis equal

plt.show()
```

```python
# ELEMENT CONNECTION
REF_R0 , NORMAL0 = ELEMENT_CONNECTION(NCONEC , KCONEC0 , MESH_0_POS , NE[0], N_STRUCT)
```

```python
N_total = N[0]
```

```python
KCONEC  = KCONEC0[0:9, :]
```

```python
# FORM SYSTEM OF EQUATIONS FOR EACH REGION
import time

tic = time.perf_counter()
H , G = GHMAT6(MESH_0_POS , NCONEC , KCONEC0 , NE[0] , N[0] , DEPTH);
toc = time.perf_counter()
print("Elapsed time:", toc - tic, "seconds")
```

```python
# FREQUENCY ITERATIONS

# Preallocate if not already done
X_axis = np.zeros(ttmax)

CR = np.zeros(ttmax)
CT = np.zeros(ttmax)
KR = np.zeros(ttmax)
KT = np.zeros(ttmax)

WAVE = {
    "k": np.zeros(ttmax),
    "lambda": np.zeros(ttmax),
    "HL": np.zeros(ttmax),
    "T": np.zeros(ttmax),
    "omega": np.zeros(ttmax),
    "Amp": np.zeros(ttmax)
}

# ============================
# MAIN LOOP
# ============================

for tt in range(ttmax):

    print(tt + 1)

    X_axis[tt] = kh[tt]

    # WAVE NUMBER
    k = kh[tt] / DEPTH

    # WAVE PERIOD
    T = np.sqrt((2*np.pi)**2 / (k * gravity * np.tanh(k * DEPTH)))

    # WAVE FREQUENCY
    f = 1.0 / T

    # ANGULAR FREQUENCY
    omega = 2*np.pi*f

    Wave_height = HL * (2*np.pi / k)
    Amp = Wave_height / 2

    # Store wave properties
    WAVE["k"][tt] = k
    WAVE["lambda"][tt] = 2*np.pi / k
    WAVE["HL"][tt] = Wave_height / (2*np.pi / k)
    WAVE["T"][tt] = T
    WAVE["omega"][tt] = omega
    WAVE["Amp"][tt] = Amp

    # SAVE MATRICES
    A = H.copy()
    B = G.copy()

    # ------------------------
    # BOUNDARY CONDITIONS
    # ------------------------
    # FREE SURFACE BC
    A = BC_FREE_SURFACE(A, B, omega, gravity, NCONEC, REF_R0["FS"], KCONEC, NE, 0)

    # FAR FIELD BC
    A = BC_FAR_FIELD(A, B, k, NCONEC,REF_R0["IN"],  KCONEC, NE, 0)
    A = BC_FAR_FIELD(A, B, k, NCONEC,REF_R0["OUT"], KCONEC, NE, 0)

    # SCATTERING BC (from incident velocity potential in the left far field)
    BC_S = BC_SCATTERING(Wave_height, k, omega, gravity, DEPTH, MESH_0_POS, NCONEC, REF_R0["IN"], KCONEC, N, NE, 0)

    # RHS VECTOR
    FX_S = FX_FUNCTION(NCONEC, B, BC_S, N, NE)

    # SOLVE SYSTEM
    Coef_S = np.linalg.solve(A, FX_S)

    # ENERGY FLUX    
    FI = (WIDTH_CHANNEL) * (1/8) * rho * gravity * (Wave_height**2) * (omega / k) * (
        0.5 * (1 + (2 * k * DEPTH) / np.sinh(2 * k * DEPTH))
     )

    FLUX_REFLE, FLUX_TRANS = ENERGY_FLUX(Coef_S, omega, rho, k, gravity, Wave_height, DEPTH, NCONEC,
        REF_R0["IN"],  MESH_0_POS,
        REF_R0["OUT"], MESH_0_POS,
        KCONEC,
        N, NE, 0, 0)

    CR[tt] = np.sqrt(FLUX_REFLE / FI)
    CT[tt] = np.sqrt(FLUX_TRANS / FI)

    KR[tt] = FLUX_REFLE / FI
    KT[tt] = FLUX_TRANS / FI

# END LOOPS
```

```python
# Bragg reflection of water waves by multiple submerged semi-circular breakwaters - Yong Liu, Hua-jun Li, Lei Zhu
kh_Liu = np.array([
0.064, 0.152, 0.263, 0.391, 0.546,
0.734, 0.925, 1.083, 1.234, 1.356,
1.511, 1.654, 1.780, 1.934, 2.126,
2.321, 2.512, 2.685, 2.869
])

CR_Liu = np.array([
0.051, 0.118, 0.194, 0.262, 0.314,
0.336, 0.320, 0.288, 0.247, 0.210,
0.163, 0.123, 0.092, 0.060, 0.030,
0.009, -0.004, 0.003, 0.006
])

CT_Liu = np.sqrt(1 - CR_Liu**2)

plt.figure(1)

# plot(kh, KT, 'bs')
# 'bs' stands for blue squares
plt.plot(kh, CR, 'b--', label='CR_BEM')
plt.plot(kh, CT, 'r--', label='CR_BEM')

plt.plot(kh_Liu, CR_Liu, 'b+', label='CR_Liu')
plt.plot(kh_Liu, CT_Liu, 'r+', label='CR_Liu')

plt.xlim([0, 3.00])
plt.ylim([0, 1.00])

plt.xlabel('kh')
plt.ylabel('Coefficient value')

plt.legend()
plt.grid(True)

plt.savefig("CR_CT.pdf")

# Optional: display the plot
plt.show()
```

```python
# VISUALIZE FREE SURFACE
# %matplotlib widget
PHI_FS = VISUAL_FREE_SURFACE(Coef_S, k, omega, gravity, Wave_height, DEPTH, NCONEC, REF_R0["FS"], MESH_0_POS, MESH_0_POS, KCONEC, N, NE, 0)
```

```python
# Boundary conditions
# Inlet  dPHI/dn + ik PHI = 0
# Outlet dPHI/dn - ik PHI = 0
# FS     dPHI/dn - omega^2/g PHI = 0
# Bottom dPHI/dn = 0
# Imper  dPHI/dn = 0
# Front  dPHI/dn = 0
# Back   dPHI/dn = 0

# Interface PHI_0 = PHI_1,  dPHI_0/dn = dPHI_1/dn

# c = 1 for internal points and c = 1/2 for boundary points where the boundary is smooth

Fluxes_S = np.zeros_like(KCONEC, dtype=np.complex128)

Fluxes_S = FLUXES_FREE_SURFACE(Coef_S, Fluxes_S, omega, gravity, NCONEC, REF_R0["FS"], KCONEC, N, NE, 0)
Fluxes_S = FLUXES_FAR_FIELD(Coef_S, Fluxes_S, k, NCONEC, REF_R0["IN"], KCONEC, N, NE, 0)
Fluxes_S = FLUXES_FAR_FIELD(Coef_S, Fluxes_S, k, NCONEC, REF_R0["OUT"], KCONEC, N, NE, 0)

CX = np.linspace(0, 3.7, 101)
CY = np.zeros_like(CX)
CZ = np.zeros_like(CX)

DSOL = INTERNAL_POINTS(Coef_S, Fluxes_S, DEPTH, CX, CY, CZ, MESH_0_POS, NCONEC, KCONEC, NE, 0) # ADD REGION
```

```python
ETA_IP = 2 * np.real((1j * omega / gravity) * DSOL * np.exp(-1j * omega * np.pi / omega))
```

```python
plt.figure(4)

plt.plot(CX, ETA_IP, 'b--', label='ETA')

plt.xlim([0, 3.70])
plt.ylim([-0.025, 0.025])

plt.xlabel('x')
plt.ylabel('eta')

plt.legend()
plt.grid(True)

# Optional: display the plot
plt.show()
```
