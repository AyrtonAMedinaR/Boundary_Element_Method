# 3D Frequency-Domain Boundary Element Method (BEM)

This tutorial demonstrates the implementation of a 3D Boundary Element Method (BEM) to analyze the interaction between regular linear water waves and submerged porous structures.

The formulation is based on:
- Linear potential flow theory
- Time-harmonic (frequency-domain) analysis
- Boundary integral equation formulation

The objective is to compute:
- Wave transmission and reflection
- Energy dissipation due to porosity

```python
# Variables of interest

# Water depth
DEPTH = 2.5

# Half wave flume width
WIDTH_CHANNEL = 10

# Cylinder dimensions
RADIO  = 1/2*DEPTH
HEIGHT = 2/3*DEPTH

#     Point 1      2             3               4
#     ↓            ↓             ↓               ↓
#     +------------+-------------+---------------+
#     |            |   Object    |               |
#     |  Incident  |   .......   |  Transmitted  |   y
#     |     →      |   .     .   |               |   ↑
#(0,0)+.....→......|......+......|...............|    → x
#     |     →      |   .     .   |               |
#     |            |   .......   |               |
#     |            |             |               |
#     +------------+-------------+---------------+

# Point 1, 2, 3 and 4
FLUME_LENGTH_X = [0, 2*DEPTH, 4*DEPTH, 6*DEPTH]

# Location of the center's structures in x and y coordinates
# The middle of the wave flume is at y=0 
LOC_STRUCT_X = [3*DEPTH]
LOC_STRUCT_Y = [0]

# Number of structures
N_struct  = len(LOC_STRUCT_X);

# Number of regions: outer and inner regions
NUM_OF_REG = 2

gravity = 9.8056
rho     = 999.1033
```

```python
# NUMBER OF FREQUENCIES
ttmax = 1

# WAVE PERIODS
kh = np.linspace(3.00, 3.00, ttmax)
# WAVE HEIGHT
HL = 0.01
# PARAMETERS FOR POROSITY
eps = 0.10

# NONLINEAR ITERATIONS
ITERATION = 1
# LINEAR INITIAL GUESS
Gporous = 0.5
```

```python
# GMSH meshes
MAIN_MESH(DEPTH,WIDTH_CHANNEL,RADIO,HEIGHT,FLUME_LENGTH_X,LOC_STRUCT_X,LOC_STRUCT_Y)
```

```python
# Process the outer region
MESH_0_POS, MESH_0_QUADS9 = PROCESS_MESH("OUTER_REGION.msh")

# Process the inner region
MESH_1_POS, MESH_1_QUADS9 = PROCESS_MESH("INNER_REGION.msh")
```

```python
# Number of nodes per quad element
NCONEC = 9;
```

```python
# PRE-ASSIGN
N  = [0] * NUM_OF_REG
NE = [0] * NUM_OF_REG
```

```python
N[0] ,  NE[0] ,  KCONEC0 = COORDINATES(MESH_0_POS , MESH_0_QUADS9)
N[1] ,  NE[1] ,  KCONEC1 = COORDINATES(MESH_1_POS , MESH_1_QUADS9)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(
    MESH_0_POS[:,0],
    MESH_0_POS[:,1],
    MESH_0_POS[:,2],
    c='k',
    s=1
)

#ax.grid(True)
ax.set_box_aspect([7.5,5,1])  # axis equal

plt.show()
```

```python
# ELEMENT CONNECTION
REF_R0 , NORMAL0 = ELEMENT_CONNECTION(NCONEC , KCONEC0 , MESH_0_POS , NE[0])
REF_R1 , NORMAL1 = ELEMENT_CONNECTION(NCONEC , KCONEC1 , MESH_1_POS , NE[1])
```

```python
# MATCHING NODES BETWEEN TWO ADJACENT REGIONS (INTERFACE)

# CYL 1
NODES_MATCH_CYL_1 = MATCHING(REF_R0['POROUS_31'] ,  MESH_0_POS  ,  KCONEC0  ,  REF_R1['POROUS_31'] ,  MESH_1_POS  ,  KCONEC1);

# CYL 2
# NODES_MATCH_CYL_2 = MATCHING(REF_R0['POROUS_32'] ,  MESH_0_POS  ,  KCONEC0  ,  REF_R1['POROUS_32'] ,  MESH_1_POS  ,  KCONEC1);
```

```python
# N0      = N[0]
N_total = N[0] + N[1]
```

```python
KCON_N_0  = KCONEC0[0:9, :]
KCON_N_01 = KCONEC1[0:9, :] + N[0]

KCONEC = np.hstack((KCON_N_0, KCON_N_01))
```

```python
# FORM SYSTEM OF EQUATIONS FOR EACH REGION
import time

tic = time.perf_counter()
H0 , G0 = GHMAT6(MESH_0_POS , NCONEC , KCONEC0 , NE[0] , N[0] , DEPTH);
toc = time.perf_counter()
print("Elapsed time:", toc - tic, "seconds")

tic = time.perf_counter()
H1 , G1 = GHMAT6(MESH_1_POS , NCONEC , KCONEC1 , NE[1] , N[1] , DEPTH);
toc = time.perf_counter()
print("Elapsed time:", toc - tic, "seconds")

# H1_matlab = np.loadtxt('H1_output.txt', delimiter=',')
# G1_matlab = np.loadtxt('G1_output.txt', delimiter=',')
# np.round(G1[:,-1],4)
# np.round(G1_matlab[:,-1],4)
```

```python
# MERGE H AND G MATRICES
H = np.zeros((N_total, N_total), dtype=np.complex128)
H[:N[0], :N[0]] = H0
H[N[0]:N_total, N[0]:N_total] = H1

# NE0 = NE[0]
NE_total = NE[0] + NE[1]

G_CONE0  = NCONEC * NE[0]
G_CONE01 = NCONEC * NE_total

G = np.zeros((N_total, G_CONE01), dtype=np.complex128)
G[:N[0], :G_CONE0] = G0
G[N[0]:N_total, G_CONE0:G_CONE01] = G1
```

```python
# READ THE POROUS NODES IN BOTH REGIONS
[STORE_NODE_R0, STORE_NODE_R1] = POROUS_NODES(NCONEC, KCONEC0, KCONEC1, MESH_0_POS, MESH_1_POS, NODES_MATCH_CYL_1);
```

```python
# FREQUENCY ITERATIONS

# Preallocate if not already done
X_axis = np.zeros(ttmax)

CR = np.zeros(ttmax)
CT = np.zeros(ttmax)
KR = np.zeros(ttmax)
KT = np.zeros(ttmax)
KL = np.zeros(ttmax)

WAVE = {
    "k": np.zeros(ttmax),
    "lambda": np.zeros(ttmax),
    "HL": np.zeros(ttmax),
    "T": np.zeros(ttmax),
    "omega": np.zeros(ttmax),
    "Amp": np.zeros(ttmax)
}

Err_1 = np.zeros((ITERATION, ttmax))

# ============================
# MAIN LOOP
# ============================

for tt in range(ttmax):

    print(tt + 1)

    X_axis[tt] = kh[tt]

    # WAVE NUMBER
    k = kh[tt] / DEPTH

    # WAVE PERIOD
    T = np.sqrt((2*np.pi)**2 / (k * gravity * np.tanh(k * DEPTH)))

    # WAVE FREQUENCY
    f = 1.0 / T

    # ANGULAR FREQUENCY
    omega = 2*np.pi*f

    Wave_height = HL * (2*np.pi / k)
    Amp = Wave_height / 2

    # Store wave properties
    WAVE["k"][tt] = k
    WAVE["lambda"][tt] = 2*np.pi / k
    WAVE["HL"][tt] = Wave_height / (2*np.pi / k)
    WAVE["T"][tt] = T
    WAVE["omega"][tt] = omega
    WAVE["Amp"][tt] = Amp

    # Wall thickness
    b = 0.0254 * 1.0

    # NONLINEAR
    l = 0.0   # b/eps if desired

    # ============================
    # NONLINEAR ITERATIONS
    # ============================

    for HHH in range(ITERATION):

        print(HHH + 1)

        # SAVE MATRICES
        A = H.copy()
        B = G.copy()

        # ------------------------
        # BOUNDARY CONDITIONS
        # ------------------------
        # FREE SURFACE BC
        A = BC_FREE_SURFACE(A, B, omega, gravity, NCONEC, REF_R0["FS"], KCONEC, NE, 0)

        # FAR FIELD BC
        A = BC_FAR_FIELD(A, B, k, NCONEC,REF_R0["IN"],  KCONEC, NE, 0)
        A = BC_FAR_FIELD(A, B, k, NCONEC,REF_R0["OUT"], KCONEC, NE, 0)

        # POROUS STRUCTURE BC
        A = BC_COMPLEX_G(A, B, k, NCONEC, STORE_NODE_R0, STORE_NODE_R1, N, NE, Gporous, 0, 1)

        # SCATTERING BC (from incident velocity potential in the left far field)
        BC_S = BC_SCATTERING(Wave_height, k, omega, gravity, DEPTH, MESH_0_POS, NCONEC, REF_R0["IN"], KCONEC, N, NE, 0)

        # RHS VECTOR
        FX_S = FX_FUNCTION(NCONEC, B, BC_S, N, NE)

        # SOLVE SYSTEM
        Coef_S = np.linalg.solve(A, FX_S)

        # UPDATE Gporous
        G_old = Gporous

        Gporous = UPDATE_G_POROUS(Coef_S, k, omega, eps, l, Gporous, NCONEC, STORE_NODE_R0, STORE_NODE_R1, MESH_0_POS, MESH_1_POS, N, 0, 1)

        # ERROR
        Err_1[HHH, tt] = abs(Gporous - G_old)

        # ENERGY FLUX    
        FI = (WIDTH_CHANNEL) * (1/8) * rho * gravity * (Wave_height**2) * (omega / k) * (
            0.5 * (1 + (2 * k * DEPTH) / np.sinh(2 * k * DEPTH))
         )

        FLUX_REFLE, FLUX_TRANS = ENERGY_FLUX(Coef_S, omega, rho, k, gravity, Wave_height, DEPTH, NCONEC,
            REF_R0["IN"],  MESH_0_POS,
            REF_R0["OUT"], MESH_0_POS,
            KCONEC,
            N, NE, 0, 0)

        CR[tt] = np.sqrt(FLUX_REFLE / FI)
        CT[tt] = np.sqrt(FLUX_TRANS / FI)
    
        KR[tt] = FLUX_REFLE / FI
        KT[tt] = FLUX_TRANS / FI
        KL[tt] = 1.0 - KR[tt] - KT[tt]        

# END LOOPS
```

```python
# figure(15)
plt.figure(15)

# plot(kh, KT, 'bs')
# 'bs' stands for blue squares
plt.plot(kh, KT, 'bs')

# xlim([0 1.5])
plt.xlim([0, 3.00])

# ylim([0 1])
plt.ylim([0, 1])

# Optional: display the plot
plt.show()
```
